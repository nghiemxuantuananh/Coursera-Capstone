## Central Toronto cordinate

```python
latitude = 43.654260
longitude = -79.360636
```

## Plotting 

```python
Toronto_map = folium.Map(location=[latitude, longitude], zoom_start=15) # generate map centred around Ecco


# add Ecco as a red circle mark
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    popup='Ecco',
    fill=True,
    color='red',
    fill_color='red',
    fill_opacity=0.6
    ).add_to(Toronto_map)


# add popular spots to the map as blue circle markers
for latitude, longitude, label in zip(df3['Latitude'], df3['Longitude'], df3['Borough']):
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        fill=True,
        color='blue',
        fill_color='blue',
        fill_opacity=0.6
        ).add_to(Toronto_map)

# display map
Toronto_map
```
